```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model 
```

```python
df = pd.read_csv("2_train.csv", delimiter='\t')
df.head(2)
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)', fontsize='14')
plt.ylabel('price(US$)', fontsize='14')
plt.scatter(df.area, df.price, color='red', marker='*')
```

```python
pd.crosstab(df.area, df.price).plot(kind='bar')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.score(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```
